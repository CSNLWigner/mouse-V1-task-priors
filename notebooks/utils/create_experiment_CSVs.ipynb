# Create Experiment CSVs
The `DATASET1_CellTrialTable.csv` is too large to open. To access the data, we need to break it into smaller CSVs.

## Imports

```python
%load_ext autoreload
%autoreload 2
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm

from utils_corbo.data_access import load_tables, retrieve_experiment_stats
from utils_corbo.data_access import query_CellTrialTable_dataframe, compute_CellTrialTable_df_stats
from utils_corbo.data_access import retrieve_experiment_ids, save_CellTrialTable_df, check_CellTrialTable_existence
```

## Create Experiment CSV workflow

```python
tables = load_tables()
experiment_ids = retrieve_experiment_ids(tables)
```

```python
for experiment_id in tqdm(experiment_ids):
    try:
        check_CellTrialTable_existence(experiment_id)
    except FileExistsError as e:
        print(f"Error: {e}")
        continue
    
    df_experiment = query_CellTrialTable_dataframe(experiment_id)
    
    experiment_stats = retrieve_experiment_stats(experiment_id, tables,)
    db_experiment_stats = compute_CellTrialTable_df_stats(df_experiment)
    assert experiment_stats == db_experiment_stats, f'Stats for {experiment_id} do not match'

    save_CellTrialTable_df(df_experiment)
```
